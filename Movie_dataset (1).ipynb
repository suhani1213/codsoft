Year with highest average voting

```python
data.columns
```

```python
# Convert 'Votes' column to numeric, handling commas and errors
data['Votes'] = pd.to_numeric(data['Votes'].str.replace(',', ''), errors='coerce')

# Calculate the mean votes per year
data.groupby('Year')['Votes'].mean()
```

```python
sns.barplot(x='Year',y='Votes',data=data)
plt.title('Average Votes per Year')
plt.show()
```

Highest Average Revenue

```python
sns.barplot(x='Year',y='Votes',data=data) # Replace 'revenue' with 'Votes'
plt.title('Average Votes per Year')
plt.show()
print(data.columns) # Print columns to verify their names
data.groupby('Year')['Votes'].mean().sort_values(ascending=False) # Use 'Votes' here as well
```
